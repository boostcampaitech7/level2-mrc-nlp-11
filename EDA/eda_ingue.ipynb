# mrc-1과 mrc-0의 개수비교

## train

```python
import pandas as pd
import re

df_train_ingue = pd.DataFrame({
    'title': dataset['train']['title'],
    'question':dataset['train']['question'],
    'answers':dataset['train']['answers'],
    'id':dataset['train']['id'],
    'document_id':dataset['train']['document_id'],
    '__index_level_0__':dataset['train']['__index_level_0__']
})

df_train_ingue
```

```python
# 'id' 컬럼에서 'mrc-1'로 시작하는 값들의 개수
mrc_1_count_train = df_train_ingue[df_train_ingue['id'].str.startswith('mrc-1')].shape[0]

# 'id' 컬럼에서 'mrc-0'로 시작하는 값들의 개수
mrc_0_count_train = df_train_ingue[df_train_ingue['id'].str.startswith('mrc-0')].shape[0]

print(f"'mrc-1' 개수: {mrc_1_count_train}, 'mrc-0' 개수: {mrc_0_count_train}")
```

```python
import matplotlib.pyplot as plt

# 데이터 정의
labels = ['mrc-1', 'mrc-0']
counts = [mrc_1_count_train, mrc_0_count_train]

# 막대 그래프 그리기
plt.figure(figsize=(6, 4))
plt.bar(labels, counts, color=['blue', 'green'])

# 그래프 제목 및 레이블 추가
plt.title("'mrc-1' and 'mrc-0' count compare")
plt.xlabel('ID type')
plt.ylabel('count')

# 개수 표시
for i, count in enumerate(counts):
    plt.text(i, count + 5, str(count), ha='center', va='bottom')

# 그래프 보여주기
plt.show()
```

```python
sns.countplot(y='title', data=df_train_ingue, order=df_train_ingue['title'].value_counts().index[:5])
```
